```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris Dataset
df = pd.read_csv("IRIS.csv")

# Create Dashboard
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Bar Chart - Species Count
df['species'].value_counts().plot(
    kind='bar',
    ax=axes[0]
)
axes[0].set_title('Species Count')
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Count')

# 2. Histogram - Sepal Length
axes[1].hist(
    df['sepal_length'],
    bins=15
)
axes[1].set_title('Sepal Length Distribution')
axes[1].set_xlabel('Sepal Length')
axes[1].set_ylabel('Frequency')

# 3. Scatter Plot
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species',
    ax=axes[2]
)
axes[2].set_title('Sepal Length vs Petal Length')

# Adjust Layout
plt.tight_layout()

# Show Dashboard
plt.show()
```
